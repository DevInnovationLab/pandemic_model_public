```python
import pandas as pd
import numpy as np

from pandemic_model.stats.reg import IntervalMixedRE, design_matrix_auto
```

```python
ptrs_raw = pd.read_csv("../../data/clean/vaccine_ptrs.csv")

encode_vars = ['has_prototype', 'pathogen', 'platform']
ptrs_encoded = pd.get_dummies(ptrs_raw, columns=encode_vars)
ptrs_encoded['respondent_enc'] = pd.Categorical(ptrs_encoded['respondent']).codes

regress_df = ptrs_encoded.drop(columns=['disease', 'respondent'])
regress_df = regress_df[regress_df[['value_min', 'value_max']].notna().all(axis=1)]
```

```python
L = regress_df['value_min'].to_numpy()
U = regress_df['value_max'].to_numpy()
group_idx = regress_df['respondent_enc'].to_numpy()
X = regress_df.drop(
	columns=['value_min', 'value_max', 'respondent_enc', 'has_prototype_0', 'has_prototype_1']
).to_numpy()
```

```python
regress_df[regress_df['respondent_enc'] == 1]
```

```python
intreg = IntervalMixedRE(family='logistic', maxiter=2)
result = intreg.fit(L, U, X, group_idx, verbose=True)
```
