```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import ipywidgets as widgets
from ipywidgets import interact
```

```python
def plot_posterior(alpha, beta, n, y):
    if y > n:
        print("Error: y cannot be greater than n.")
        return
    
    alpha_prime = alpha + y
    beta_prime = beta + (n - y)
    
    theta = np.linspace(0, 1, 1000)
    prior = stats.beta.pdf(theta, alpha, beta)
    posterior = stats.beta.pdf(theta, alpha_prime, beta_prime)
    
    plt.figure(figsize=(8, 5))
    plt.plot(theta, prior, label=f'Prior: Beta({alpha}, {beta})', linestyle='dashed')
    plt.plot(theta, posterior, label=f'Posterior: Beta({alpha_prime}, {beta_prime})', linewidth=2)
    
    plt.xlabel(r'\(\theta\)', fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.title('Beta Prior and Posterior Distribution', fontsize=14)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
interact(plot_posterior, 
         alpha=widgets.BoundedFloatText(min=0.1, max=100, step=0.1, value=2.0, description='Alpha'),
         beta=widgets.BoundedFloatText(min=0.1, max=100, step=0.1, value=2.0, description='Beta'),
         n=widgets.BoundedIntText(min=1, max=100, step=1, value=10, description='n'),
         y=widgets.BoundedIntText(min=0, max=100, step=1, value=5, description='y'))
```

```python
def plot_surveillance_posteriors(false_rate=0.3, n_pathogens=100, n_signals=10, k=1.0):
    """
    Visualizes the prior and posterior distributions from a two-stage surveillance system.
    
    Args:
        false_rate (float): Base false positive rate used for prior (default 0.3)
        n_pathogens (int): Number of cases to simulate (default 100)
        k (float): Factor to scale alpha and beta parameters (default 1.0)
    """
    # Set up prior parameters with k scaling
    alpha = (1 - false_rate) * k
    beta = false_rate * k
    n = n_signals  # signals per round
    prior_mean = alpha / (alpha + beta)
    
    # Generate random true/false cases
    is_false = np.random.binomial(1, false_rate, n_pathogens).astype(bool)
    
    # Generate surveillance signals following MATLAB logic
    p1 = 0.4 + 0.2 * ~is_false  # First round probabilities
    p2 = 0.2 + 0.6 * ~is_false  # Second round probabilities
    
    y1 = np.random.binomial(n, p1)  # First round signals
    y2 = np.random.binomial(n, p2)  # Second round signals
    
    # Calculate posteriors
    posterior1 = (alpha + y1) / (n + k)
    posterior2 = (alpha + y2) / (n + k)
    posterior1_mean_true = np.mean(posterior1)
    posterior2_mean_true = np.mean(posterior2)
    
    # Plot distributions
    plt.figure(figsize=(15, 5))
    
    # Prior distribution
    plt.subplot(1, 3, 1)
    theta = np.linspace(0, 1, 1000)
    prior = stats.beta.pdf(theta, alpha, beta)
    plt.plot(theta, prior, 'k-', lw=2)
    plt.axvline(prior_mean, color='black', linestyle='--', label=f'Prior mean: {prior_mean:.3f}')
    plt.title(f'Prior Distribution\n(α={alpha:.2f}, β={beta:.2f})')
    plt.xlabel('θ')
    plt.ylabel('Density')
    plt.xlim([0, 1])
    plt.legend()
    
    # First round posteriors
    plt.subplot(1, 3, 2)
    plt.hist(posterior1, bins=25, density=True, stacked=True)
    
    # Add vertical lines for means
    plt.axvline(prior_mean, color='black', linestyle='--', label=f'Prior mean: {prior_mean:.3f}')
    plt.axvline(posterior1_mean_true, color='blue', linestyle='--', label=f'Posterior mean: {posterior1_mean_true:.3f}')
    
    plt.title(f'First Round Posterior Distribution\n(α={alpha:.2f}, β={beta:.2f})')
    plt.xlabel('Posterior probability')
    plt.ylabel('Density')
    plt.xlim([0, 1])
    plt.legend()
    
    # Second round posteriors
    plt.subplot(1, 3, 3)
    plt.hist(posterior2, bins=25, density=True, stacked=True)
    
    # Add vertical lines for means
    plt.axvline(prior_mean, color='black', linestyle='--', label=f'Prior mean: {prior_mean:.3f}')
    plt.axvline(posterior2_mean_true, color='blue', linestyle='--', label=f'Posterior mean: {posterior2_mean_true:.3f}')
    
    plt.title(f'Second Round Posterior Distribution\n(α={alpha:.2f}, β={beta:.2f})')
    plt.xlabel('Posterior probability')
    plt.ylabel('Density')
    plt.xlim([0, 1])
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Create interactive widget
interact(plot_surveillance_posteriors,
        false_rate=widgets.FloatSlider(min=0.1, max=0.9, step=0.1, value=0.3, 
                                     description='False positive rate'),
        n_pathogens=widgets.IntSlider(min=100, max=20000, step=100, value=10000,
                                description='Pathogen arrivals'),
        n_signals=widgets.IntSlider(min=1, max=50, step=1, value=10, description='Num signals'),
        k=widgets.IntSlider(min=1, max=200, step=1, value=1,
                                   description='K factor'))
```
